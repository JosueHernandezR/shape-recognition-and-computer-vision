## Problema 1: Umbralizaci칩n de una im치gen

```python
# Importaciones para la resoluci칩n de este problema
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Leyendo imagen
imagen_covid = cv2.imread('./Resources/imageCovid.jpg')
#Poniendo color RGB
covid_rgb = cv2.cvtColor(imagen_covid, cv2.COLOR_BGR2RGB)

#Graficando la imagen
plt.figure(dpi=100)
plt.title("RGB")
plt.imshow(covid_rgb)
plt.show()
```
